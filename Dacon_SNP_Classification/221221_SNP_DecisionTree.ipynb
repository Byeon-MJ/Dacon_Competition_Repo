```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import os

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import preprocessing
```

```python
train = pd.read_csv('/content/gdrive/MyDrive/tmp/dataset/dacon_SNP/train.csv')
test = pd.read_csv('/content/gdrive/MyDrive/tmp/dataset/dacon_SNP/test.csv')
info = pd.read_csv('/content/gdrive/MyDrive/tmp/dataset/dacon_SNP/snp_info.csv')
```

```python
def get_x_y(df):
    if 'class' in df.columns:
        df_x = df.drop(columns=['id', 'class'])
        df_y = df['class']
        return df_x, df_y
    else:
        df_x = df.drop(columns=['id'])
        return df_x
```

```python
train_x, train_y = get_x_y(train)
test_x = get_x_y(test)
```

```python
train_x
```

```python
train_y
```

### Label Encoding

```python
class_le = preprocessing.LabelEncoder()
snp_le = preprocessing.LabelEncoder()
snp_col = [f'SNP_{str(x).zfill(2)}' for x in range(1,16)]
```

```python
snp_col
```

```python
train_data = []
for col in snp_col:
    train_data += list(train_x[col].values)
```

```python
train_y = class_le.fit_transform(train_y)
snp_le.fit(train_data)
```

```python
train_x.columns
```

```python
for col in train_x.columns:
    if col in snp_col:
        train_x[col] = snp_le.transform(train_x[col])
        test_x[col] = snp_le.transform(test_x[col])
```

```python
train_x
```

```python
train_x, val_x, train_y, val_y = train_test_split(
    train_x, train_y, test_size=0.4, random_state = 100)
```

# Model Fit

```python
# max_depth=depth
model = DecisionTreeClassifier(random_state=100, max_depth=6)
model.fit(train_x, train_y)
```

```python
train_pred = model.predict(train_x)
val_pred = model.predict(val_x)
test_pred = model.predict(test_x)
```

```python
print(f'Train Score : {accuracy_score(train_y, train_pred)}')
print(f'Validation Score : {accuracy_score(val_y, val_pred)}')
```

## 하이퍼 파라미터 튜닝

```python
def test_depth(depth):
    model = DecisionTreeClassifier(random_state=100, max_depth=depth)
    model.fit(train_x, train_y)
    train_pred = model.predict(train_x)
    val_pred = model.predict(val_x)
    print(f"depth : {depth}")
    print(f'Train Score : {accuracy_score(train_y, train_pred)}')
    print(f'Validation Score : {accuracy_score(val_y, val_pred)}')
```

```python
for i in range(1, 20):
    test_depth(i)
```

## Tree Visualization
